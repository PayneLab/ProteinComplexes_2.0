```python
pcutoff = 0.05
results_data = pd.DataFrame(columns=['Gene', 'Protein', 'Test_Type', 'p_value'])
protein_list = sub_20s
for gene in test_genes:
    mutation_df = colon.get_mutations().loc[colon.get_mutations()['Gene'] == gene]
    #mutation_df = mutation_df.loc[mutation_df['Mutation'].str.contains('Silent')]
    missense_truncation_groups = {'frameshift substitution': 'T', 
                'frameshift deletion': 'T', 'frameshift insertion': 'T', 
                'stopgain': 'T', 'stoploss': 'T', 'nonsynonymous SNV': 'M',
                'nonframeshift insertion': 'M','nonframeshift deletion': 'M', 
                'nonframeshift substitution': 'M'}
    mutation_df = mutation_df.replace(missense_truncation_groups)
    mutation_df = mutation_df.join(colon.get_proteomics()[protein])[['Gene', 'Mutation', protein]]
    mutation_df['Mutation'] = mutation_df['Mutation'].astype(pd.api.types.CategoricalDtype(categories = ['T', 'M'], ordered = True))
    mutation_df = mutation_df.sort_values('Mutation').groupby('Sample_ID').first()[['Mutation']]
    pd.set_option('display.max_rows', 10)
    #print(mutation_df)

    source_data = colon.get_clinical()[['Patient_ID']]
    source_data['Sample_ID'] = source_data.index

    proteomics_data = colon.get_proteomics()[protein_list]
    source_data = source_data.join(proteomics_data)

    # Extract all the normal matched samples into a new df and remove the 'N' to match them with tumor samples
    normal = source_data.copy()
    normal = normal.loc[normal['Patient_ID'].str.contains('N')]
    normal['Patient_ID'] = normal['Patient_ID'].str.replace('N', '')
    normal.index = normal['Patient_ID']

    # Remove the normal samples from the tumor df
    tumor = source_data.loc[~source_data['Patient_ID'].str.contains('N')]
    tumor.index = tumor['Patient_ID']

    # Remove unmatched samples from each dataframe
    tumor = tumor.loc[tumor['Patient_ID'].isin(normal['Patient_ID'])]
    normal = normal.loc[normal['Patient_ID'].isin(tumor['Patient_ID'])] 
    test_data = tumor.copy()[['Sample_ID', 'Patient_ID']]

    for protein in protein_list:
        test_data[protein] = (tumor[protein] + 10) - (normal[protein] + 10)

    test_data.index = test_data['Sample_ID']
    test_data = test_data.join(mutation_df)
    test_data['Mutation'].fillna('WT', inplace = True)

    make_simple_boxplot(test_data, gene)
    
    wt = test_data.loc[test_data['Mutation'] == 'WT']
    all_mutations = test_data.loc[test_data['Mutation'] != 'WT']
    missense = test_data.loc[test_data['Mutation'] == 'M']
    truncation = test_data.loc[test_data['Mutation'] == 'T']
    
    for protein in protein_list:
        # Add WT vs All Mutated t-test
        if len(wt[protein].dropna(axis = 0)) > 2 and len(all_mutations[protein].dropna(axis = 0)) > 2:
            p_val_wt_all = scipy.stats.ttest_ind(wt[protein].dropna(axis = 0), all_mutations[protein].dropna(axis = 0))[1]
            if p_val_wt_all <= pcutoff:
                results_data = results_data.append({
                    'Gene': gene,
                    'Protein': protein,
                    'Test_Type': 'WT_All_Mutated',
                    'p_value': p_val_wt_all
                }, ignore_index = True)
        
        # Add WT vs M t-test
        if len(wt[protein].dropna(axis = 0)) > 2 and len(missense[protein].dropna(axis = 0)) > 2:
            p_val_wt_m = scipy.stats.ttest_ind(wt[protein].dropna(axis = 0), missense[protein].dropna(axis = 0))[1]
            if p_val_wt_m <= pcutoff:
                results_data = results_data.append({
                    'Gene': gene,
                    'Protein': protein,
                    'Test_Type': 'WT_Missense',
                    'p_value': p_val_wt_m
                }, ignore_index = True)
        
        # Add WT vs T t-test
        if len(wt[protein].dropna(axis = 0)) > 2 and len(truncation[protein].dropna(axis = 0)) > 2:
            p_val_wt_t = scipy.stats.ttest_ind(wt[protein].dropna(axis = 0), truncation[protein].dropna(axis = 0))[1]
            if p_val_wt_t <= pcutoff:
                results_data = results_data.append({
                    'Gene': gene,
                    'Protein': protein,
                    'Test_Type': 'WT_Truncation',
                    'p_value': p_val_wt_t
                }, ignore_index = True)


print(results_data)
```

```python
def make_simple_boxplot(plot_data, gene):
    a4_dims = (100, 50)
    fig, ax = plt.subplots(figsize=a4_dims)
    sns.set(font_scale = 5)
    plot_data = pd.melt(plot_data, id_vars = ['Sample_ID', 'Patient_ID', 'Mutation'], var_name = 'Protein', value_name = 'T-N')
    boxplt = sns.boxplot(data = plot_data, x = 'Protein', y = 'T-N', hue = 'Mutation', showfliers = False)
    boxplt = sns.stripplot(data = plot_data, x = 'Protein', y = 'T-N', hue = 'Mutation', dodge = True, jitter = True, color = '.3')
    boxplt.set_title(gene + ' mutations', fontsize = '200')
    handles, labels = ax.get_legend_handles_labels()
    l = plt.legend(handles[0:3], labels[0:3])
```

### The Proteasome

### Proteasomal Proteins
Perform t-tests on tumor vs normal protein expression for the proteins within the proteasomal complexes

```python
proteasome_proteins = subunitNames['PA700 complex'] + subunitNames['26S proteasome'] + subunitNames['20S proteasome'] + subunitNames['PA28-20S proteasome'] + subunitNames['PA700-20S-PA28 complex'] + subunitNames['PA28gamma-20S proteasome']
sig_results = []
# Remove duplicates
proteasome_proteins = list(set(proteasome_proteins))
# Bonferroni correction of the p-value
alpha = 0.05 / len(proteasome_proteins)
# Perform t-tests
for protein in proteasome_proteins:
    pval = ttest_proteomics(protein, cancer_type = 'colon')
    if pval < alpha:
        sig_results.append(protein)
        print('p-value for ' + protein + ' (tumor vs normal): ' + str(pval))
```

### Divide significant results into subunits of the proteasome

```python
sub_20s = []
sub_19s = []
pa28 = []
for protein in sig_results:
    if 'PSMA' in protein or 'PSMB' in protein: sub_20s.append(protein)
    elif 'PSMC' in protein or 'PSMD' in protein: sub_19s.append(protein)
    else: pa28.append(protein)  
sub_20s.sort()
sub_19s.sort()
pa28.sort()
```
