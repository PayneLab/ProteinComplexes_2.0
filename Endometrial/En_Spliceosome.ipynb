## The Spliceosome
The spliceosome is an extremely intricate complex involving hundreds of proteins and five snRNAs. Hundreds of these proteins exhibit differential expression in tumor vs normal cells. In order to make analysis easier, we're going to use spliceosome_proteins.csv downloaded from http://spliceosomedb.ucsc.edu/ (the Spliceosome Database) PMID: 23118483, which divides spliceosomal proteins based on which complex they are a part of (A, B, B*, etc.). The complexes represent the components of the spliceosome at each step of the splicing process.

### Perform t-tests on all proteins in the spliceosome

```python
%%capture --no-stdout --no-display
# The line above will capture stderr and prevent warnings from cluttering our data

subunits = pd.read_csv('Data_Files/spliceosome_proteins.csv')[['Gene Symbol', 'Complex', 'Class / family']]
spliceosome_proteins = list(set(subunits['Gene Symbol']))
sig_results = []
# Remove duplicates
spliceosoame_proteins = list(set(spliceosome_proteins))
# Bonferroni correction of the p-value
alpha = 0.05 / len(spliceosome_proteins)
# Perform t-tests
for protein in spliceosome_proteins:
    pval = ttest_proteomics(protein, cancer_type = 'en')
    if pval < alpha:
        sig_results.append(protein)
        print('p-value for ' + protein + ' (tumor vs normal): ' + str(pval))
```

```python
print(str(len(sig_results)) + '/' + str(len(spliceosome_proteins)) + ' spliceosomal proteins are significantly different')
```

### Obtain a list of proteasome complexes with proteins that exhibit differential expression between tumor and normal cells
This turns out to be all of the sub-complexes in the database

```python
splice_complexes = subunits.loc[subunits['Gene Symbol'].isin(sig_results)]
splice_complexes = tidy_split(splice_complexes, 'Complex', sep = ', ')
print(list(set(splice_complexes['Complex'])))
```

SRSF1 is an SR protein (has an RNA recognition motif; preferentially selects splice sites on the pre-mRNA strands). It is frequently overexpressed in tumors. It is a transcriptional target of MYC and is also regulated in response to splicing factor Sam68. Sam68 (KHDRBS1) switches splicing of the SRSF1 transcript from a nonsense-mediated decay targeted isoform to the major, translatable isoform. Results in an increase in SRSF1 levels.
