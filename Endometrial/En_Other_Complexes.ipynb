## GSK3B-HSP90AA1-PKM2 complex
#### Not present in normal samples --> likely functional in tumor samples
GSK3B mediates Thr-328 phosphorylation of PKM2 (an isoenzyme of pyruvate kinase that is expressed in cells with a high rate of nucleic acid synthesis, especially tumor cells) induced by HSP90. This phosphorylation maintains PKM2 stability and its functions in proliferation, etc. Associated with cancer (PMID: 29262861).  GSK3B is known to be overexpressed in ovarian cancer (PMID: 18606491) Note: We have no proteomics/phosphorylation data on PKM2

Plot of the ratio of GSK3B / HSP90AA1. No normal samples have any GSK3B, whereas most tumor samples show expression

```python
plot_ratios('GSK3B', 'HSP90AA1', by_patient = True)
```

5 cancer patients presented with mutations in transcription factors that regulate GSK3B (unknown activating/repressing)

```python
gsk_tf_list = ['CDX2', 'CREB1', 'NFKB1', 'TFCP2']
plot_ratios('GSK3B', 'HSP90AA1', mutation_list = gsk_tf_list)
```
